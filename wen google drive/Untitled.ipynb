```python
# pydrive
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
import pandas as pd

# Auth at first time
gauth = GoogleAuth()
gauth.LocalWebserverAuth()
```

```python


def ListFolder(parent, path = ''):
    filelist=[]
    col = ['資源標題','年級','科目/單元','資源連結','mimeType',
        'path', 'user', 'user_email', 'time', 'file_id']
    df = pd.DataFrame(columns=col)
    
    drive = GoogleDrive(gauth)
    file_list = drive.ListFile({'q': "'%s' in parents and trashed=false" % parent}).GetList()
    for f in file_list:
        #print(f['title'])
        if f['mimeType']=='application/vnd.google-apps.folder': # if folder
            #filelist.append({"id":f['id'],"title":f['title'],"list":ListFolder(f['id'])})
            df_child = ListFolder(f['id'], path + '/' + f['title'])
            df = pd.concat([df, df_child], axis=0, ignore_index=True)
            print(f['title'])
        else:
            filelist.append(f['title'])
            #title
            #alternateLink
            #id
            #createdDate/ modifiedDate
            #ownerNames
            print(' - ', f['title'], ': ', path)
            '''
            print(f['alternateLink'])
            print(f['mimeType'])
            print(f['id'])
            print(f['createdDate'])
            print(f['owners'][0]['displayName'])
            print(f['owners'][0]['emailAddress'])
            '''
            df = df.append({'資源標題':f['title'],
                            '年級':'',
                            '科目/單元':'',
                            '資源連結':f['alternateLink'],
                            'mimeType':f['mimeType'],
                            'path':path,
                            'user':f['owners'][0]['displayName'],
                            'user_email':f['owners'][0]['emailAddress'],
                            'time':f['createdDate'],
                            'file_id':f['id']
                            },ignore_index=True)
            #print(f)
            
    return df
    


df = ListFolder('185rkewkiXC3zCs0woDN6ERabXGFg9kEC')

df.to_csv('output.csv', index=False)
display(df.head())
```
