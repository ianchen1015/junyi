```python
import pandas as pd
from IPython.display import display

output_df = pd.read_csv('output.csv')
crawler_df = pd.read_csv('fb_crawler_output.csv')

for label in ['post_link', 'user_name', 'user_link', 'upload_time']:
    output_df[label] = ''
    
display(output_df.head())
display(crawler_df.head())
```

```python
for i in range(len(output_df['資源標題'])):
    #print(i)
    find = 0
    for j in range(len(crawler_df['file_name'])):
        if output_df['資源標題'][i] == crawler_df['file_name'][j]:
            find = 1
            for label in ['post_link', 'user_name', 'user_link', 'upload_time']:
                output_df[label][i] = crawler_df[label][j]
    if find == 0:
        print('cannot find : ', output_df['資源標題'][i])
        
display(output_df.head())
```

```python
output_df.to_csv('combine_output.csv', index=False)
```
