```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep
```

```python

url = 'https://www.facebook.com/groups/247170855413829/files/'
usr = "ianchen1015@gmail.com"
pwd = "minecraft20"

#driver.quit()
driver = webdriver.Chrome(executable_path=r'./chromedriver')

def login():
    driver.get('https://www.facebook.com/')
    print ("Opened facebook")
    sleep(1)

    username_box = driver.find_element_by_id('email')
    username_box.send_keys(usr)
    print ("Email Id entered")
    sleep(1)

    password_box = driver.find_element_by_id('pass')
    password_box.send_keys(pwd)
    print ("Password entered")

    login_box = driver.find_element_by_id('loginbutton')
    login_box.click()

    print ("Login Done\n")

login()
driver.get(url)
sleep(3)
```

```python
cnt = 1
while True:
    try:
        load_more_btn = driver.find_elements_by_xpath('//a[@class="pam uiBoxLightblue uiMorePagerPrimary"]')[0]
        load_more_btn.click()
        sleep(5)
        print(cnt)
        cnt += 1
    except:
        break


print("Finished")
```

```python
pageSource = driver.page_source

soup = BeautifulSoup(pageSource, "html5lib")
li_lst = soup.findAll('li',{'class':'_pu_'})

col = ['file_name','post_link','user_name','user_link','upload_time']
df = pd.DataFrame(columns=col)

#li = li_lst[0]
for li in li_lst:
    li_a = li.findAll('a')

    # title
    a_title = li_a[0]
    file_name = a_title.get_text().split('.')[0]
    post_link = 'https://www.facebook.com' + a_title['href']

    # user
    a_user = li_a[1]
    user_name = a_user.get_text()
    user_link = a_user['href']

    # time
    a_time = li_a[2]
    upload_time = a_time.get_text()

    #print('{}\n{}\n{}\n{}\n{}\n'.format(file_name, post_link, user_name, user_link, upload_time))
    print(file_name)

    df = df.append({'file_name':file_name,
                    'post_link':post_link,
                    'user_name':user_name,
                    'user_link':user_link,
                    'upload_time':upload_time
                    },ignore_index=True)
df.to_csv('fb_crawler_output.csv', index=False)
df.head()
```
